```python
from transformers import pipeline
zero_shot_classifier=pipeline("zero-shot-classification")
text=['A person got killed.He was a young boy.He had many enemies.The accused was his enemy.']
final_classification=zero_shot_classifier(text,candidate_labels=['Accused will win the case','Accused will lose the case'])
final_classification[0]['labels'][0]
```
